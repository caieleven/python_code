# Applied Machine Learning with Python
## Code 3.2 Regularization Techniques 


```python
from scipy import stats
import numpy as np
import pandas as pd 
import matplotlib
import matplotlib.pyplot as plt
import math
from sklearn.linear_model import Lasso 
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
```

### A. Generating Data

#### Training data

```python
x = np.linspace(0,100,30)
train  = pd.DataFrame(data={'x1': x,'x2': x**2, 'x3': x**3, 'x4': x**4})
```

#### Testing data

```python
x = np.linspace(100,200,30)
test = pd.DataFrame(data={'x1': x,'x2': x**2, 'x3': x**3,'x4': x**4})
```

#### Function f(x)

```python
def f(x):
        return pd.DataFrame(3*x.x1 - 1.3*x.x2)
```

### B. Linear Model

```python
linear_model = LinearRegression(normalize=False)
linear_model.fit(train,f(train))
linear_model.coef_
```

Plotting

```python
plt.plot(test.x1, linear_model.predict(test),'b',alpha=0.6, label='Prediction')
plt.plot(test.x1, f(train),'r--', label='Data')
plt.legend(loc='best', frameon=False)
plt.xlabel('x1')
plt.ylabel('y')
plt.show()
```

### C. Lasso Model

```python
lasso_model = Lasso(alpha=0.5,normalize=True, max_iter=1e6)
lasso_model.fit(train,f(train))
```

Plotting

```python
plt.plot(test.x1, lasso_model.predict(test),'b',alpha=0.6, label='Prediction')
plt.plot(test.x1, f(test),'r--', label='Data')
plt.legend(loc='best', frameon=False)
plt.xlabel('x1')
plt.ylabel('y')
plt.show()
```

```python
lasso_model.coef_
```

### D. Ridge Model

```python
ridge_model = Ridge(alpha=0.5)
ridge_model.fit(train, f(train)) 
```

```python
plt.plot(test.x1, ridge_model.predict(test),'b',alpha=0.6, label='Prediction')
plt.plot(test.x1, f(test),'r--', label='Data')
plt.legend(loc='best', frameon=False)
plt.xlabel('x1')
plt.ylabel('y')
plt.show()
```
